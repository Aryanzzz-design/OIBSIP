```python
import pandas as pd

df = pd.read_csv('/content/AB_NYC_2019.csv')

print("First 5 rows of the dataset:")
display(df.head())

print(f"\nNumber of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
})

print("Missing values in each column:")
display(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False))
```

```python
# Create a copy of the DataFrame to perform cleaning operations
df_cleaned = df.copy()
```

### Handling Missing Values

*   **`name` and `host_name`**: These columns have a very small number of missing values (0.03% and 0.04% respectively). Dropping these rows will have a minimal impact on the dataset size while ensuring data integrity for these key identifiers.
*   **`reviews_per_month`**: This column has a significant number of missing values (over 20%). It is reasonable to assume that a missing value here indicates no reviews for that listing, so we will fill these `NaN` values with `0`.
*   **`last_review`**: This date column also has over 20% missing values, correlating with `reviews_per_month`. When converted to a datetime type, `NaN` values will naturally become `NaT` (Not a Time), which is a suitable representation for an absent date in a datetime column.

```python
# Create a copy of the DataFrame to perform cleaning operations
df_cleaned = df.copy()

# Drop rows where 'name' or 'host_name' are missing
initial_rows = df_cleaned.shape[0]
df_cleaned.dropna(subset=['name', 'host_name'], inplace=True)
rows_dropped_name_host = initial_rows - df_cleaned.shape[0]
print(f"Dropped {rows_dropped_name_host} rows due to missing 'name' or 'host_name'.")

# Fill missing 'reviews_per_month' with 0
df_cleaned['reviews_per_month'].fillna(0, inplace=True)

# Convert 'last_review' to datetime. Missing values will become NaT.
df_cleaned['last_review'] = pd.to_datetime(df_cleaned['last_review'])
```

After these operations, let's re-check the missing values and the DataFrame information to confirm the changes.

```python
print("Missing values after cleaning:")
missing_after_cleaning = df_cleaned.isnull().sum()
display(missing_after_cleaning[missing_after_cleaning > 0])

print("\nDataFrame info after cleaning:")
df_cleaned.info()
```

```python
print("First 5 rows of the cleaned dataset:")
display(df_cleaned.head())


print(f"\nNumber of rows in cleaned dataset: {df_cleaned.shape[0]}")
print(f"Number of columns in cleaned dataset: {df_cleaned.shape[1]}")
```

```python
initial_rows_before_dropping_duplicates = df_cleaned.shape[0]

duplicate_rows = df_cleaned.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_rows}")

# Drop duplicate rows
df_cleaned.drop_duplicates(inplace=True)

rows_dropped_duplicates = initial_rows_before_dropping_duplicates - df_cleaned.shape[0]
print(f"Dropped {rows_dropped_duplicates} duplicate rows.")
print(f"Number of rows after removing duplicates: {df_cleaned.shape[0]}")
```

### Data Standardization

Standardization is a data preprocessing technique used to scale numerical features so that they have a mean of 0 and a standard deviation of 1. This is particularly useful for many machine learning algorithms that are sensitive to the scale of input features.

We will apply standardization to the numerical columns of the `df_cleaned` DataFrame to ensure consistent scaling.

```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns for standardization
numerical_cols = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Exclude 'id' and 'host_id' as they are identifiers and not features to be scaled
# 'number_of_reviews' and 'calculated_host_listings_count' are counts and can be scaled
numerical_cols_to_scale = [col for col in numerical_cols if col not in ['id', 'host_id']]

print(f"Numerical columns identified for standardization: {numerical_cols_to_scale}")

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the selected numerical columns
df_cleaned[numerical_cols_to_scale] = scaler.fit_transform(df_cleaned[numerical_cols_to_scale])

print("\nFirst 5 rows of the DataFrame after standardization of numerical columns:")
display(df_cleaned[numerical_cols_to_scale].head())
```

```python
# Identify numerical columns for outlier detection (excluding identifiers and 'last_review')
numerical_cols_for_outliers = [col for col in df_cleaned.select_dtypes(include=['int64', 'float64']).columns if col not in ['id', 'host_id']]

print("Outlier Detection using IQR Method:")

outlier_summary = {}
for col in numerical_cols_for_outliers:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]
    num_outliers = outliers.shape[0]

    if num_outliers > 0:
        outlier_summary[col] = num_outliers
        print(f"  Column '{col}': {num_outliers} outliers found.")

if not outlier_summary:
    print("  No significant outliers detected using the IQR method for the selected numerical columns.")

# Note: This step only identifies outliers. Further action (e.g., capping, removal) would depend on the specific analysis goals.
```
