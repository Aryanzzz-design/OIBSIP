## Customer Segmentation

```python
# Select features for segmentation
# These features represent spending habits, purchase behavior, and some demographic info
features = [
    'Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits',
    'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmpOverall', 'Age',
    'Customer_Days', 'MntTotal', 'MntRegularProds'
]

df_segmentation = df[features].copy()

display(df_segmentation.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Handle potential missing values in selected features by filling with the mean
# (though df.info() showed no missing values, it's good practice for general cases)
for col in df_segmentation.columns:
    if df_segmentation[col].isnull().any():
        df_segmentation[col].fillna(df_segmentation[col].mean(), inplace=True)

# Scale the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_segmentation)

# Convert the scaled array back to a DataFrame for easier inspection
df_scaled = pd.DataFrame(df_scaled, columns=features)
display(df_scaled.head())
```

### Determine Optimal Number of Clusters (Elbow Method)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calculate sum of squared distances for different numbers of clusters
ssd = []
range_n_clusters = range(1, 11) # Test 1 to 10 clusters

for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10) # n_init to suppress warning
    kmeans.fit(df_scaled)
    ssd.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, ssd, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Distances (SSD)')
plt.grid(True)
plt.show()
```

### Apply K-Means Clustering

```python
# Apply K-Means with the chosen number of clusters (e.g., 4 from elbow method)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_scaled)

# Add the cluster labels to the original (unscaled) segmentation DataFrame
df_segmentation['Cluster'] = clusters

display(df_segmentation.head())
```

### Visualize Customer Segments using PCA

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensions for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Create a DataFrame for the PCA results
df_pca = pd.DataFrame(data = df_pca, columns = ['PCA_Feature1', 'PCA_Feature2'])
df_pca['Cluster'] = clusters

display(df_pca.head())
```

```python
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PCA_Feature1', y='PCA_Feature2', hue='Cluster', data=df_pca, palette='viridis', s=100, alpha=0.7)
plt.title('Customer Segments (PCA-reduced data)')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```

### Analyze Cluster Characteristics (Bar Charts)

```python
# Calculate the mean of each feature for each cluster
cluster_means = df_segmentation.groupby('Cluster')[features].mean()
display(cluster_means)
```

```python
# Select a few key features to visualize for clearer insights
selected_features_for_bar_chart = [
    'Income', 'MntWines', 'MntMeatProducts', 'NumStorePurchases', 'Kidhome', 'Age'
]

# Create bar charts for each selected feature across clusters
plt.figure(figsize=(18, 12))
for i, feature in enumerate(selected_features_for_bar_chart):
    plt.subplot(2, 3, i + 1) # 2 rows, 3 columns
    sns.barplot(x=cluster_means.index, y=feature, data=cluster_means, palette='viridis')
    plt.title(f'Average {feature} by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel(f'Average {feature}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Insights and Recommendations

```python
for cluster_num in cluster_means.index:
    print(f"\n--- Cluster {cluster_num} ---")
    cluster_data = cluster_means.loc[cluster_num]

    print("**Characteristics:**")
    if cluster_data['Income'] > df_segmentation['Income'].mean():
        print(f"- High Income (Avg. ${cluster_data['Income']:.2f})")
    else:
        print(f"- Lower Income (Avg. ${cluster_data['Income']:.2f})")

    if cluster_data['MntWines'] > df_segmentation['MntWines'].mean():
        print(f"- High Spender on Wines (Avg. ${cluster_data['MntWines']:.2f})")
    else:
        print(f"- Lower Spender on Wines (Avg. ${cluster_data['MntWines']:.2f})")

    if cluster_data['MntMeatProducts'] > df_segmentation['MntMeatProducts'].mean():
        print(f"- High Spender on Meat (Avg. ${cluster_data['MntMeatProducts']:.2f})")
    else:
        print(f"- Lower Spender on Meat (Avg. ${cluster_data['MntMeatProducts']:.2f})")

    if cluster_data['NumStorePurchases'] > df_segmentation['NumStorePurchases'].mean():
        print(f"- Frequent Store Shopper (Avg. {cluster_data['NumStorePurchases']:.2f} purchases)")
    else:
        print(f"- Less Frequent Store Shopper (Avg. {cluster_data['NumStorePurchases']:.2f} purchases)")

    if cluster_data['Kidhome'] > 0 and cluster_data['Teenhome'] > 0:
        print(f"- Has both Kids and Teens at home (Avg. Kids: {cluster_data['Kidhome']:.2f}, Teens: {cluster_data['Teenhome']:.2f})")
    elif cluster_data['Kidhome'] > 0:
        print(f"- Has Kids at home (Avg. Kids: {cluster_data['Kidhome']:.2f})")
    elif cluster_data['Teenhome'] > 0:
        print(f"- Has Teens at home (Avg. Teens: {cluster_data['Teenhome']:.2f})")
    else:
        print("- No Kids or Teens at home")

    print("**Recommendations:**")
    if cluster_data['Income'] > df_segmentation['Income'].mean() and cluster_data['MntWines'] > df_segmentation['MntWines'].mean():
        print("- Offer premium products and exclusive wine club memberships.")
    elif cluster_data['Kidhome'] > 0 or cluster_data['Teenhome'] > 0:
        print("- Focus on family-friendly offers and bulk discounts.")
    else:
        print("- Tailor promotions based on their specific spending patterns (e.g., if high meat, offer meat bundles).")
```
